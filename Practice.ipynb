```python
from splinter import Browser
from bs4 import BeautifulSoup as soup
from webdriver_manager.chrome import ChromeDriverManager
```

```python
#set up Splinter
executablePath = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executablePath, headless = False)
```

```python
#visit the quote website
url = 'http://quotes.toscrape.com/'
browser.visit(url)
```

```python
html = browser.html
htmlSoup = soup(html, 'html.parser')
htmlSoup
```

```python
title = htmlSoup.find('h2').text
title
```

```python
#scrape top ten tags
tagBox = htmlSoup.find('div', class_ = 'tags-box')

tags = tagBox.find_all('a', class_ = 'tag')

for tag in tags:
    word = tag.text
    print(word)
```

```python
texts = htmlSoup.find_all('div', class_='quote')

for text in texts:
    print(text.text)
```

```python
for x in range(1,6):
    html = browser.html
    quoteSoup = soup(html, 'html.parser')
    quotes = quoteSoup.find_all('span', class_ = 'text')
    for quote in quotes:
        words = quote.text
        print("page ", x, " -------------")
        print(words)
    browser.links.find_by_partial_text('Next').click
```

```python
url = 'http://books.toscrape.com/'
html = browser.html
clean = soup(html, 'html.parser')
titles = clean.find_all('h3')

for title in titles:
    print(title.text)
```
